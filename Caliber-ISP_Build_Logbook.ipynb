### Logbook Entry 1: Motor Mount Tube and Centering Ring Assembly

Today I began work on the motor mount assembly for my L1 certification rocket. The main task was to epoxy the centering rings onto the motor mount tube. For this step, I used a basic department store two-part epoxy with an estimated 15-minute working time. This provided enough time to mix the epoxy, apply it carefully, and position the parts before the adhesive began to set.

I installed two centering rings onto the motor tube. Each centering ring was positioned so that the outer face of the ring sat approximately 1/8 inch from the end of the motor tube. This was done on both ends of the tube, meaning each ring was slightly inset from its respective end. I used this spacing to help keep the assembly clean and consistent while still leaving enough bonding area between the motor tube and the centering rings.

After reviewing the parts again, I noticed that there was a third centering ring included with the kit. I appear to have skipped this third ring because it was not clearly shown or called out in the main assembly steps of the instruction manual I was following. Since this is my first high-power certification build, I am trying to document these details carefully so I can review the structure and make sure the final assembly is safe and appropriate before flight.

Overall, the two installed centering rings appear to be bonded securely to the motor tube. The epoxy was applied around the joint between each centering ring and the motor tube, creating a solid adhesive connection. I will need to confirm the purpose and placement of the third centering ring before proceeding too far with the build, especially since the motor mount is a critical load-bearing part of the rocket.


## Sunday, June 14, 2026

### Logbook Entry 2: Installing the Motor Mount Assembly into the Fuselage

Later in the day, I continued the build by installing the completed motor mount and centering ring assembly into the 34-inch fuselage section. This was one of the more important structural steps in the build because the motor mount needs to be aligned properly and securely bonded to the inside of the airframe. A poor bond in this area could affect thrust transfer, recovery loads, and overall flight reliability.

To prepare the joint, I applied a continuous bead of epoxy around the inside of the 34-inch fuselage section where the centering rings would contact the airframe. I also applied epoxy around the outside radius of the centering rings themselves. This created adhesive contact on both mating surfaces before the assembly was inserted into the fuselage. The goal was to form a two-surface epoxy bond rather than relying on epoxy applied to only one side of the joint.

Once the epoxy was applied, I inserted the motor mount assembly into the fuselage and seated it into position. After setting the assembly, I stood the fuselage section upright with the motor mount facing downward and touching the flat surface it was standing on. This helped ensure that the rear of the motor mount was flush with the end face of the fuselage section. Keeping the motor mount flush is important because it affects the final motor position, aft-end alignment, and how future components such as motor retention may fit.

The assembly was left upright while the epoxy began to cure. This setup allowed gravity to help keep the motor mount seated in the correct position while the adhesive set. From inspection, the motor mount appeared to be aligned and properly seated against the bottom surface. This step completed the initial installation of the motor mount into the main fuselage section, although I will still need to inspect the cured joint and confirm whether any additional fillets or reinforcement are needed.

## Monday, June 15, 2026

### Logbook Entry: Fin Preparation and Initial Fin Installation

Today I worked on preparing the fins for installation into the pre-slotted airframe. The first step was sanding each fin down to a smooth finish. This removed rough edges and surface imperfections from the fin material and made the fins easier to inspect before bonding. I also rounded off the leading edge of each fin to remove the sharp square edge left from manufacturing and create a cleaner fin profile.

After smoothing the fins, I test fit the fin tabs into the pre-slotted airframe. This was done to check how each fin seated into its slot and to make sure the tabs fit correctly before applying epoxy. Test fitting was important because once epoxy is mixed and applied, there is limited time to make adjustments. Verifying the fit beforehand helped reduce the chance of alignment issues during the final installation.

I also sanded the tab edge that would be epoxied during the fin installation. This was done to improve the bonding surface between the fin tab and the internal motor mount structure. A slightly roughened surface gives the epoxy more area to grip compared to a smooth surface, which should help create a stronger mechanical bond.

I originally considered laser cutting a fin alignment jig out of scrap wood to hold the fins at 90 degrees from each other, but I decided to scrap that idea for this step. Instead, I epoxied two fins in place and used Scotch tape to create tension across the airframe. The tape acted as a simple clamp to hold the fins in position while the epoxy began to cure. This method allowed the fins to stay seated in the slots without needing a separate jig.

Only two fins were epoxied in place during this work session. I chose to install them in stages so I could focus on alignment and fitment without rushing the process. Once the epoxy has cured, I will inspect the two installed fins for straightness, bond quality, and overall alignment before continuing with the remaining fins.

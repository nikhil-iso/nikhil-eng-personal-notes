## USST Rocketry: Boat tail manufacturing Notes
---
### Info
Author: Nikhil Patel  
Context: This notebook serves as a logbook for USST propulsion work

## June 11, 2026

Today I removed the boat tail shell from the two-piece press mold. The mold was 3D printed in PETG and designed by Nishok Deenadayalan. The mold consists of a positive inner mold, which the composite material is wrapped around, and a negative outer shell, which is pressed over the layup to squeeze and evenly distribute the epoxy/resin.

After allowing the composite part to cure for 48 hours, I attempted to remove it from the mold. My first approach was to heat the external face using a heat gun, with the goal of activating the mold release and making the part easier to remove.

This unintentionally caused the PETG to warp. The heat gun produced enough localized heat to soften and partially melt the surface of the plastic. After noticing this, I changed the removal strategy and accepted that the mold would likely need to be sacrificed.

I then used a pry bar and gentle force to separate the mold from the composite part. During this process, the PETG mold cracked in several areas due to the prying force. This likely happened because the mold release on the positive mold was not fully activated, which prevented the part from releasing cleanly.

After a long removal process, the PETG mold broke into many small pieces, along with several larger shards and sections. However, the composite boat tail shell was successfully removed.

The finished part had a smooth outer surface due to the press mold. The inner surface was also symmetrical and maintained the desired shape well.

### Lessons Learned

For the next attempt, I would spend more time preparing the mold surface before the layup. The positive mold should be coated more evenly and thoroughly with a much larger amount of mold release. There is very little downside to using extra release agent compared to the difficulty of removing a poorly released part.

I would also improve the fiberglass layup process. Instead of using shredded non-woven fiberglass placed in patches, I would prefer to use full sheets of fiberglass wrapped around the mold. The shredded fiberglass approach was used on Project UP for Launch Canada 2025, but after recovery the part split into shards and became flaky during removal. Using continuous sheets of fiberglass should help produce a stronger and more consistent composite shell.

For future manufacturing, I would also consider whether a different resin or epoxy system could reduce mold damage or make demolding easier. The current resin and hardener combination may have contributed to the aggressive bonding between the composite and the PETG mold.
